```python
#reading NYPD felonies Data
NYPDfelony = pd.read_csv("Data/NYPD_7_Major_Felony_Incidents.csv")
```

```python
#checking Dimensions
print("the dimensions of the dataset: ", NYPDfelony.shape)
NYPDfelony = NYPDfelony.rename(columns = {'Location 1' : 'geometry'}) #renaming column name
print(NYPDfelony.shape)
NYPDfelony.head()
```

```python
#making coloumns of latitude and longitude from column Geometry
extractZip = NYPDfelony['geometry'].str.split(',', expand=True)
extractZip.rename(columns={0: 'lat', 1: 'lon'}, inplace = True)
extractZip['lat'] = extractZip['lat'].str.split('(', expand=True)[1]
extractZip['lon'] = extractZip['lon'].str.split(')', expand=True)[0]
extractZip['OBJECTID'] = NYPDfelony.OBJECTID
extractZip.head()
```

```python
#merging extracted lat lon with original Felony Data
NYPDfelony_df = pd.merge(NYPDfelony, extractZip, on = 'OBJECTID')
print(NYPDfelony_df.columns)
NYPDfelony_df.head()
```

```python
#Converting Lat Lon to float
NYPDfelony_df.lat = NYPDfelony_df.lat.astype(float)
NYPDfelony_df.lon = NYPDfelony_df.lon.astype(float)
NYPDfelony_df.head()
```

```python
#converting NYPD Felony dataframe to geodataframe for spatial join with census tract data
from shapely.geometry import Point

crs = None
geometry = [Point(xy) for xy in zip(NYPDfelony_df.lon, NYPDfelony_df.lat)]
geo_NYPDfelony = GeoDataFrame(NYPDfelony_df, crs=crs, geometry=geometry)
geo_NYPDfelony.crs = fiona.crs.from_epsg(4326)
```
